## If-Statements with User-Defined Functions

```python
# Define a user create function for prompting price range

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else:
        return 'Not enough data'
```

```python
# Create new column based on user-defined function

df['price_range'] = df.apply(price_label, axis=1)
```

## Creating busiest hours of the day column (Task 4)

```python
# Task 4: Identify busiest hours

df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Task 4: Create a dataframe using the value count results above for descriptive stats to identify basis for bucketing

df_ord_hr = df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False).rename_axis('hour').reset_index(name='order_counts')
```

```python
# Task 4: Check new dataframe

df_ord_hr
```

```python
# Task 4: Run descriptive stat using the dataframe above

df_ord_hr['order_counts'].describe()
```

```python
# Task 4: Made a decision to use +/- one std. as the basis for bucketing
# Create list, which can be added to the dataframe create above

ord_hr_bkt = []

for orco in df_ord_hr['order_counts']:
    if orco > (1.350202e+06 + 1.078369e+06):
        ord_hr_bkt.append('Most orders')
    elif orco < (1.350202e+06 - 1.078369e+06):
        ord_hr_bkt.append('Fewest orders')
    else: ord_hr_bkt.append('Average orders')
```

```python
# Task 4: Check bucket list

ord_hr_bkt
```

```python
# Task 4: Add the list as a column to the dataframe created for descriptive stats

df_ord_hr['busy_hour'] = ord_hr_bkt
```

```python
# Task 4: Check the updated dataframe for descriptive stats

df_ord_hr
```

```python
# Task 4: Adding busiest_hour_of_day column to the intial dataframe (df_ords_prods_merged)

busiest_hr = []

for time in df_ords_prods_merged['order_hour_of_day']:
    if time in (10,11,14,15,13,12,16,9):
        busiest_hr.append('Most orders')
    elif time in (0,1,5,2,4,3):
        busiest_hr.append('Fewest orders')
    else: busiest_hr.append('Average orders')
```

```python
# Task 4: Checking the list created for the initial dataframe

busiest_hr
```

```python
# Task 4: Adding the list created as a column to the df_ords_prods_merged

df_ords_prods_merged['busiest_period_of_day'] = busiest_hr
```

```python
# Task 4: Check head after update

df_ords_prods_merged.head()
```

```python
# Task 5: Print the frequency of the new column

print(df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False))
```
