# 03. Merging (Task 5, 6)

```python
# Task 5: Test match between orders_products_combined and products

pd.merge(df_ords_prods_comb, df_prods, on = 'product_id', how = 'outer', indicator = True)
```

```python
# Task 5: Count test match types

pd.merge(df_ords_prods_comb, df_prods, on = 'product_id', how = 'outer', indicator = True)['_merge'].value_counts()
```

Task 5: CP: Based on the result above, while the majority of records between two dataframe matches, there are certain records that are only available in only one dataframe. 'Left_only' means that the product id listed in the combined orders dataframe is not available in the products dataframe, meaning that some of these orders may have referenced either (1) incorrect products or (2) decomissioned products that are no longer available in the current products list. 'Right_only' means that the product is available in the curring products list, but they were never ordered.

I initially thought of doing a left join, using the orders dataframe as the primary, but ended up deciding to conduct an inner join instead due to the immateriality of the left_only (orders only) product_ids (less than 0.1%) and to facilitate further analyses using a fully merged dataframe.

```python
# Task 5: Inner merge dataframes

df_ords_prods_merged = df_ords_prods_comb.merge(df_prods, on = 'product_id', indicator = True)
```
